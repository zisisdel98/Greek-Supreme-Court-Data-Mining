# TF-IDF Αναπαράσταση

```python
from sklearn.feature_extraction.text import TfidfVectorizer
custom_stopwords = [
    "και", "του", "της", "το", "την", "των", "σε", "με", "από", "για", "κατά","κατα","ως", "έως", "ή", "αν", "θα", "να",
    "δεν", "που", "έχει", "έχουν", "είναι", "ήταν", "στο", "στη", "στις", "στον", "αυτή", "αυτό", "αυτά", "εκ",
    "ότι", "προς", "οι", "ο", "η", "ένα", "ενα", "ενώ", "έπειτα", "επειδή", "πριν", "μετά", "όμως", "αλλά", "λοιπόν",
    ".", ",", "–", "-", "«", "»", "…", "!", ";", ":", "?", "(", ")", "=", "''", "``", "'",'αλλα', 'αυτα', 'αυτη', 'αυτο', 'ειναι', 'ενω', 'επειδη', 'επειτα', 'εχει',
    'εχουν', 'εως', 'ηταν', 'λοιπον', 'μετα', 'ομως', 'ουτε', 'οχι',
    "απο", "δε", "επί", "επι", "εξ", "δια", "εν", "πλην", "μη", "μην", "ούτε", "ναι", "όχι","παρ","οτι","τα","το","κατ","τον","στην"
]
df_clean = df.dropna(subset=['summary']).copy()

vectorizer = TfidfVectorizer(
    analyzer='word',
    stop_words=custom_stopwords,
    token_pattern=r'\b[^\d\W]+\b',       # μόνο λέξεις
    strip_accents='unicode',             # αφαιρεί τόνους
    lowercase=True,
    max_df=0.95,
    min_df=5,
    max_features=5000,
    ngram_range=(1, 2)
)
X = vectorizer.fit_transform(df_clean['summary'])  # df_clean έχει summary και case_category
```

```python
#! pip install optuna
import optuna
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans


def objective(trial):
    k = trial.suggest_int('k', 2, 40)
    model = KMeans(n_clusters=k, n_init=10, random_state=42)
    labels = model.fit_predict(X)
    score = silhouette_score(X, labels)
    return score

# Optuna Study
study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=20)

# Βέλτιστο Κ
best_k = study.best_params['k']
print(f"✅ Βέλτιστο K: {best_k} με Silhouette: {study.best_value:.4f}")

```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import metrics

# Clustering για πολλά K και Οπτικοποίηση
kappas = list(range(2, 40))
sscores = []

for k in kappas:
    model = KMeans(n_clusters=k, n_init=10, random_state=42)
    labels = model.fit_predict(X)
    score = silhouette_score(X, labels, sample_size=1000, random_state=42)
    sscores.append(score)

# Οπτικοποίηση
plt.figure(figsize=(8, 5))
plt.plot(kappas, sscores, marker='o')
plt.axvline(best_k, color='red', linestyle='--', label=f'Best k = {best_k}')
plt.xlabel('Αριθμός Clusters (K)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score για διάφορα K')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

print(f"TF-IDF repre.Τελικό Βέλτιστο K: {best_k} (score = {max(sscores):.4f})")

```

```python
import pandas as pd
from gensim.models.fasttext import load_facebook_model
from gensim.utils import simple_preprocess

stopwords = [
    # Άρθρα – οριστικά και αόριστα
    "ο", "οι", "η", "το", "οι", "τα", "τον", "την", "τους", "τις", "των",
    "ένας", "μία", "ένα", "ενας", "μια", "ενα",

    # Αντωνυμίες
    "αυτός", "αυτή", "αυτό", "αυτοί", "αυτές", "αυτά",
    "αυτος", "αυτη", "αυτο", "αυτοι", "αυτες", "αυτα",
    "εγώ", "εσύ", "εμείς", "εσείς", "εγω", "εσυ", "εμεις", "εσεις",
    "μου", "σου", "του", "της", "μας", "σας", "τους", "τους",

    # Συνδέσμοι
    "και", "ή", "ήταν", "ηταν", "αλλά", "αλλα", "όμως", "ομως", "ενώ", "ενω", "λοιπόν", "λοιπον",

    # Προθέσεις / επιρρήματα
    "σε", "με", "χωρίς", "κατά", "κατα", "από", "απο", "ως", "έως", "προς", "εν", "δια", "επι", "επί", "εκ", "εξ",
    "πριν", "μετά", "πλεον", "πλέον", "πλην", "πανω", "κάτω", "κατω", "πάνω",

    # Βοηθητικά / Modal / Ρήματα
    "είναι", "ήταν", "ήμουν", "ειναι", "ηταν", "ημουν",
    "έχει", "έχουν", "έχω", "εχει", "εχουν", "εχω",
    "θα", "να", "αν", "μη", "μην", "δεν", "δε", "ας",

    # Άλλες συχνές
    "που", "ότι", "οτι", "για", "πως", "ποιος", "ποια", "ποιο",
    "τι", "ποιοι", "ποιες", "ποια", "πώς", "ποιος", "κάποιος", "καποιος","κατ","αρθρο","άρθρο"

    # Επιφωνήματα / σύνδεση προτάσεων
    "ναι", "όχι", "οχι", "λοιπόν", "δηλαδή", "δηλαδη",

    # Αριθμοί (λόγια/γραμματικά)
    "πρώτος", "πρωτος", "δεύτερος", "δευτερος", "τρίτος", "τριτος", "τέταρτος", "τεταρτος", "ένατος",
    "πρώτη", "πρωτη", "τρίτη", "τριτη", "δεύτερη", "δευτερη", "όγδοη", "ογδοη",

    # Ποσοτικά / χρονικά
    "όσο", "οσο", "τόσο", "τοσο", "πιο", "λιγότερο", "λιγοτερο", "περισσότερο", "περισσοτερο",

    # Άλλα πολύ συχνά νομικά αδιάφορα
    "επίσης", "καθώς", "σύμφωνα", "συμφωνα", "όταν", "οταν", "αντί", "αντι"
]



# Φόρτωση pre-trained full FastText μοντέλου (.bin)
model = load_facebook_model('cc.el.300.bin')  # ← χρησιμοποίησε το .bin αρχείο

# Καθαρισμός και tokenization κειμένων
df_clean = df.dropna(subset=['summary']).copy()

# Νέο tokenize με stopword filtering
def tokenize(text):
    tokens = simple_preprocess(text, deacc=True)
    return [w for w in tokens if w not in stopwords]

tokenized_docs = df_clean['summary'].apply(tokenize).tolist()

# Ενημέρωση λεξιλογίου και fine-tuning
model.build_vocab(tokenized_docs, update=True)  # update=True → δεν πετά το παλιό vocab
model.train(tokenized_docs, total_examples=len(tokenized_docs), epochs=20)




```

```python
# πιο κοντινές λέξεις στο "ποινή"
print(model.wv.most_similar("ποινή", topn=10))

# Μέσος όρος λέξεων ανά έγγραφο (document embedding)
import numpy as np

def document_vector(doc):
    words = [word for word in doc if word in model.wv]
    return np.mean(model.wv[words], axis=0) if words else np.zeros(model.vector_size)

X = np.array([document_vector(doc) for doc in tokenized_docs])
print("X shape:", X.shape)  # (num_docs, 300)
```

```python
import optuna
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans

def objective(trial):
    k = trial.suggest_int('k', 5, 100, step=1)

    model = KMeans(n_clusters=k, n_init=10, random_state=42)
    labels = model.fit_predict(X)

    score = silhouette_score(X, labels)
    return score

study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=30)

best_k = study.best_params['k']
print(f"✅ Βέλτιστο K από Optuna: {best_k} με Silhouette: {study.best_value:.4f}")
```

```python
from sklearn.decomposition import PCA
#  PCA – μείωση από 300 → π.χ. 100 διαστάσεις
pca = PCA(n_components=100, random_state=42)
X_pca = pca.fit_transform(X)

print("X_pca shape:", X_pca.shape)  # (num_docs, 100)
```

```python
import optuna
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
from optuna.samplers import RandomSampler

sampler = RandomSampler(seed=42)

def objective(trial):
    k = trial.suggest_int('k', 5, 40, step=1)

    model = KMeans(n_clusters=k, n_init=10, random_state=42)
    labels = model.fit_predict(X_pca)

    score = silhouette_score(X_pca, labels)
    return score

study = optuna.create_study(direction="maximize",sampler=sampler)
study.optimize(objective, n_trials=20)

best_k = study.best_params['k']
print(f"✅ Βέλτιστο K από Optuna: {best_k} με Silhouette: {study.best_value:.4f}")
```



*   Embeddings + PCA + MACRO silhouette



```python
import optuna
from sklearn.metrics import silhouette_samples
from sklearn.cluster import KMeans
from optuna.samplers import RandomSampler
import numpy as np

sampler = RandomSampler(seed=42)

def objective(trial):
    k = trial.suggest_int('k', 5, 40, step=1)

    model = KMeans(n_clusters=k, n_init=10, random_state=42)
    labels = model.fit_predict(X_pca)

    # silhouette scores per sample
    s_scores = silhouette_samples(X_pca, labels)

    # macro average: mean silhouette per cluster, then mean of means
    macro_means = []
    for cluster_id in range(k):
        cluster_s = s_scores[labels == cluster_id]
        if len(cluster_s) > 0:
            macro_means.append(np.mean(cluster_s))

    macro_score = np.mean(macro_means)
    return macro_score

# Study with macro
study = optuna.create_study(direction="maximize", sampler=sampler)
study.optimize(objective, n_trials=20)

best_k = study.best_params['k']
print(f"✅ Βέλτιστο K με βάση Macro Silhouette: {best_k} με Macro Score: {study.best_value:.4f}")
```



*   MICRO ΚΑΙ MACRO SILHOUETTES



```python
from sklearn.metrics import silhouette_samples

# s(i) για κάθε δείγμα
model = KMeans(n_clusters=best_k, n_init=10, random_state=42)
labels = model.fit_predict(X_pca)

all_silhouettes = silhouette_samples(X_pca, labels)

# Silhouette Micro (μέσος όρος όλων των s(i))
sil_micro = np.mean(all_silhouettes)

# Silhouette Macro (μέσος όρος ανά cluster)
cluster_means = []
for cluster_id in range(best_k):
    cluster_s = all_silhouettes[labels == cluster_id]
    if len(cluster_s) > 0:
        cluster_means.append(np.mean(cluster_s))
sil_macro = np.mean(cluster_means)

print(f"Silhouette Micro: {sil_micro:.4f}")
print(f"Silhouette Macro: {sil_macro:.4f}")
```
